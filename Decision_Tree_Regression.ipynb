## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Height_Weight.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,-1].values.reshape(-1,1)
print(input)
output = data.iloc[:,1:2].values
print(output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt
plt.xlabel("WEIGHT")
plt.ylabel("HEIGHT")
plt.scatter(input,output)
```

## ***TRAINING THE DATASET***

```python
from sklearn.tree import DecisionTreeRegressor

Decision_Tree_Regressor_Model = DecisionTreeRegressor()

Decision_Tree_Regressor_Model.fit(input,output)
```

## ***PREDICTING THE OUTPUT USING THE TRAINED MODEL***

```python
X = 134
X = np.reshape(X,(-1,1))

print(Decision_Tree_Regressor_Model.predict(X),"IS THE HEIGHT OF THE PERSON WITH WEIGHT",X)
```

## ***VISUALIZING THE RESULT OF THE TRAINED MODEL***

```python
import matplotlib.pyplot as plt

X_val = np.arange(min(input),max(input),0.01)
X_val = X_val.reshape(len(X_val),1)
plt.xlabel("WEIGHT")
plt.ylabel("HEIGHT")
plt.scatter(input,output)
plt.plot(X_val,Decision_Tree_Regressor_Model.predict(X_val),color="red")
```
